# Q5

```python
splitdate=youtube_data["publish_time"].str.split("-",n=3,expand=True)
splitdate
```

```python
youtube_data['year']=splitdate[0]
youtube_data
```

```python
yearyoutube_data=youtube_data.loc[(youtube_data.year=="2017")|(youtube_data.year=="2018")]
yearyoutube_data
```

```python
yearyoutube_datamean=yearyoutube_data.groupby(["year","category_id"])[["likes"]].mean()
yearyoutube_datamean
```

```python
datalikes=yearyoutube_datamean.pivot_table(index="category_id",columns="year",values="likes")
datalikes
```

# Q6

```python
datalikesn=datalikes[["2018"]]
datalikesn
```

```python
dataf=datalikesn.sort_values("2018",ascending=False).head(3).reset_index()
dataf.rename(columns={"2018":"likes"},inplace=True)
dataf["2018"]=[2018,2018,2018]
dataf
```

```python
x=sns.barplot(data=dataf,x="2018",y="likes",hue="category_id")
```
